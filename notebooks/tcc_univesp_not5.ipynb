## 3.0 Data Analysis

### 3.1 Casos de Dengue

```python
mask = df_analysis['Classificação final do caso'] == 1
dftemp = df_analysis[mask]
```

```python
df = dftemp.groupby(['Ano da notificação'])['Classificação final do caso'].sum().reset_index()
plt.figure()
sns_plot = sns.barplot(data=df,x='Ano da notificação', y='Classificação final do caso')
plt.grid()
plt.xlabel('[Ano]')
plt.ylabel('[Números de Caso de Dengue]')
plt.title('Casos no Estado de São Paulo')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.show()
fig = sns_plot.get_figure()
fig.savefig('%s/analysis_full.png'%(locsave),dpi=400,bbox_inches='tight')
plt.close()
```

```python
columns_plot = dftemp.columns[4:-1]
ncolumns_plot = len(columns_plot)
```

```python
for k1 in range(0,ncolumns_plot):

    colname = columns_plot[k1]
    df = dftemp.groupby(['Ano da notificação',colname])['Classificação final do caso'].sum().reset_index()
    plt.figure()
    sns_plot = sns.barplot(data=df,x='Ano da notificação', y='Classificação final do caso',hue=colname)
    plt.grid()
    plt.xlabel('[Ano]')
    plt.ylabel('[Números de Caso de Dengue]')
    plt.title('Casos no Estado de São Paulo \n Segmentação: %s'%colname)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    plt.show()
    fig = sns_plot.get_figure()
    fig.savefig('%s/analysis%d.png'%(locsave,k1),dpi=400,bbox_inches='tight')
    plt.close()
```
